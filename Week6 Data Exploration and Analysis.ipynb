Non-Normal Data Vs Normal Data Overview

    Data that has a normal distribution has most of its data at the center.  Normally distributed data is also a symmetric spread of data.  Normal distribution has many uses in social sciences, and nature.  The mean, median, and mode are exactly the same.  The two values that describe a normal distribution are the standard deviation and the frequency. 

    The empirical rule is used to easily describe probability values within various standard deviations.  The probability values 68%, 95%, and 99.7% describe one, two, and three standard deviations from the mean with the empirical rule.  The empirical rule is a quick way to get an overview of any outliers in the data.


    Non-normal distributions can lack symmetry, have extreme values, and take other shapes that are different from normal distributed data.  A few non normal distributions include, beat, exponential, gamma, inverse, poisson and skewed distributions.  Many data sets fit a non normal model  There are several tests where normality can be assumed for non normal data such as, sample Z test, Ttest, and ANOVA that can be used with large data samples.


    A few reasons for non normal data sets are outliers, multiple distributions, insufficient data, and poor graphing.  These reasons can give a normal shaped data set many non normal characteristics.  A normal distributed data set will most likely have less issues and difficulties then a non normal dat set.

# Creating data for the age variable that is normally distributed. 

````python
data = np.random.normal(loc=45, scale=3, size=1000)
data_round =np.round(data)
df_age = pd.DataFrame(data_round, columns=['age'])
 

    
````

````python
pre_df['age'] = df_age
````

# Makeing fifty random outliers with two for loops.

````python
import random
df_age_out = df_age

#adding in 50 outliers randomly
for x in range(25):
    r_int = random.randint(1,999)
    r_age = random.randint(1,20)
    df_age.at[r_int,'age']=r_age

for x in range(25):
    r_int = random.randint(1,999)
    r_age = random.randint(60,80)
    df_age.at[r_int,'age']=r_age
pre_df['age'] = df_age_out
````
